Program ini dibuat untuk melakukan validasi data untuk memastikan kualitas data sudah sesuai dengan kriteria yang diinginkan.

```python
# Import libraries
import pandas as pd
import great_expectations as ge
```

```python
# Read csv file to dataframe
df = pd.read_csv('P2M3_clarence_manzo_data_clean.csv')

# Convert to Great Expectations dataframe
gdf = ge.from_pandas(df)
gdf
```

```python
# 1. Expect SKU to be unique
gdf.expect_column_values_to_be_unique("sku")
```

```python
# 2. Expect lead_time to be between 1 and 30
gdf.expect_column_values_to_be_between("lead_time", min_value=1, max_value=30)
```

```python
# 3. Expect routes to be in Route A, B, or C
gdf.expect_column_values_to_be_in_set("routes", ["Route A", "Route B", "Route C"])
```

```python
# 4. Expect price to be of type float
gdf.expect_column_values_to_be_of_type("price", "float")
```

```python
# 5. Expect SKU to have increasing index
gdf.expect_column_values_to_match_regex("sku", r"^SKU\d+$")
```

```python
# 6. Expect price to be minimum around 1
gdf.expect_column_min_to_be_between("price", min_value=1, max_value=2)
```

```python
# 7. Expect revenue to be greater than costs
gdf.expect_column_pair_values_A_to_be_greater_than_B("revenue_generated","costs")
```
